**4. Countries in WHO region**

```python
import plotly.express as px

whoRegionCountries_df = df.groupby(['WHO Region'])[['Country/Region']].count()
whoRegionCountries_df.columns = ['Total Countries']  # Rename the column directly
whoRegionCountries_df.sort_values(by='Total Countries', ascending=False, inplace=True)  # Sort the DataFrame in place

fig = px.pie(whoRegionCountries_df, values='Total Countries', names=whoRegionCountries_df.index, title='Countries in WHO Region', hole=0.5)
fig.show()
```

```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt


# Selecting relevant features for clustering
features = ['Deaths', 'Recovered', 'Confirmed']

# Dropping rows with missing values
df_clustering = df[features].dropna()

# Standardizing the data
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df_clustering)

# Choosing the number of clusters
num_clusters = 5

# Building the K-means model
kmeans = KMeans(n_clusters=num_clusters, n_init=10, random_state=42)  # Explicitly setting n_init
kmeans.fit(scaled_data)

# Assigning clusters to the data points
df_clustering['cluster'] = kmeans.labels_

# Visualizing the clusters (example: 2D plot)
plt.scatter(df_clustering['Deaths'], df_clustering['Recovered'], c=df_clustering['cluster'], cmap='tab10')  # Different color scheme
plt.xlabel('Deaths')
plt.ylabel('Recovered')
plt.title('K-means Clustering of COVID-19 Data')
plt.show()

# You can also explore the characteristics of each cluster
cluster_centers = scaler.inverse_transform(kmeans.cluster_centers_)
cluster_df = pd.DataFrame(cluster_centers, columns=features)
print(cluster_df)
```

**In unsupervised learning, such as clustering, we typically don't have ground truth labels to calculate accuracy like we do in supervised learning. However, we can evaluate the quality of clustering using metrics such as the silhouette score**

```python
from sklearn.metrics import silhouette_score

# Calculate silhouette score
silhouette_avg = silhouette_score(scaled_data, df_clustering['cluster'])
print("Silhouette Score:", silhouette_avg)
```
